```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pulsar.csv', header=-1)
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(28,25))
ax.set_ylim(-5,100)
n_lines = df.shape[0]
x = range(df.shape[1])

for row in df.iterrows():
    line = row[1].values/3 + (n_lines - row[0])
    ax.plot(x, line, lw=4, c='white', alpha=1, zorder=row[0]/n_lines)
    ax.fill_between(x, -5,line,  facecolor='black', zorder=row[0]/n_lines)

ax.set_yticks([])
ax.set_xticks([])
ax.set_xlim(min(x)-100, max(x)+100)
ax.set_ylim(-20, 110)
plt.savefig('plot.png')
```
